```python
'''
Ball and Beam Stabilization By Means of the Inverse Lyapunov Approach
'''

%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
from underactuated import PlanarRigidBodyVisualizer

import ball_and_beam_2d

# Initial Position Of The Ball
x0 = [0.45, 0.55, 0, 0] # [theta, r, theta_dot, r_dot]
#x0 = [0.35, 0.3, 0.18, 0.18] 
#x0 = [0, 0, 0, 0]
#x0 = [0.15, 0.15, 0, 0]
#x0 = [-0.45, -0.55, 0, 0]
#x0 = [-0.45, -0.69, 0, 0] # |r| = 0.70 too large, breaks controller
#x0 = [0.45, 0.6, 0, 0] # goes crazy with this i.c.
#x0 = [0.55, 0.55, 0, 0]

# Simulation
ballandbeam, controller, state_log = ball_and_beam_2d.Simulate2dBallAndBeam(x0 = x0, duration=10)

# Visualization
viz = PlanarRigidBodyVisualizer(ballandbeam, xlim=[-2, 2], ylim=[-1, 4])
viz.fig.set_size_inches(10, 5)
ani = viz.animate(state_log, 30, repeat=True)
plt.close(viz.fig)
HTML(ani.to_html5_video()) # This needs to be the last line for the video to display
```

```python
'''
Plot traces of certain states
'''

plt.figure().set_size_inches(10, 5)
plt.plot(state_log.sample_times(), state_log.data()[0, :])
plt.grid(True)
plt.legend(["theta"])

plt.figure().set_size_inches(10, 5)
plt.plot(state_log.sample_times(), state_log.data()[1, :])
plt.grid(True)
plt.legend(["r"])

plt.figure().set_size_inches(10, 5)
plt.plot(state_log.sample_times(), state_log.data()[2, :])
plt.grid(True)
plt.legend(["theta_dot"])

plt.figure().set_size_inches(10, 5)
plt.plot(state_log.sample_times(), state_log.data()[3, :])
plt.grid(True)
plt.legend(["r_dot"])
```
